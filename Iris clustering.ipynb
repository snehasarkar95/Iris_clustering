# K-Means Clustering

```python
select=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]
df1=df[select]    
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from yellowbrick.cluster import SilhouetteVisualizer
from matplotlib.lines import Line2D
```

```python
wcss=[]
k=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df1)
    wcss.append(km.inertia_)
    k.append(i)
    print(f"completed{i}/10: wcss---> {km.inertia_}")
```

```python
plt.figure(figsize=(8,5))
plt.plot(k,wcss,label="inertia")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.legend()
plt.show()
```

```python
SK=range(2,11)
sil_score=[]
for i in SK:
    km=KMeans(n_clusters=i,init="k-means++",random_state=100)
    km.fit(df1)
    labels=km.labels_
    score=metrics.silhouette_score(df1,labels,metric="euclidean",random_state=200)
    sil_score.append(score)
```

```python
sil_centers=pd.DataFrame({"Clusters":SK,
                         "Sil Score":sil_score})
sil_centers
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 100, n_init = 10, random_state = 100)
df1['cluster'] = kmeans.fit_predict(df1)
```

```python
df1.head()
```

```python
centroids = kmeans.cluster_centers_
cen_x = [i[0] for i in centroids] 
cen_y = [i[1] for i in centroids]
```

```python
df1['cen_x'] = df1.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2]})
df1['cen_y'] = df1.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2]})
```

```python
colors = ['#008080', '#FF6347', '#E50000']
df1['color'] = df1.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]})
```

```python
df1.head()
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))
# plot data
plt.scatter(df1["SepalLengthCm"], df["SepalWidthCm"], c=df1["color"], alpha = 0.8, s=20)
# plot centroids
plt.scatter(cen_x, cen_y, marker='o', c=colors, s=110)
# plot lines
for idx, val in df1.iterrows():
    x = [val["SepalLengthCm"], val["cen_x"]]
    y = [val["SepalWidthCm"], val["cen_y"]]
    plt.plot(x, y, c=val["color"], alpha=0.2)
# legend
cent_leg = [Line2D([0], [0], marker='o', color='w', label='Centroid - C{}'.format(i+1), 
            markerfacecolor=mcolor, markersize=10) for i, mcolor in enumerate(colors)]
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Cluster {}'.format(i+1), 
                   markerfacecolor=mcolor, markersize=5) for i, mcolor in enumerate(colors)]
legend_elements.extend(cent_leg)
plt.legend(handles=legend_elements, loc='upper right', ncol=2)

# title and labels
plt.title('Cluster based Sepal Length and Sepal Width\n', loc='center', fontsize=15)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```
